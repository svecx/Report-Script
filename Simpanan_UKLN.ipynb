```python
# Hitung bulan lalu (format nama bulan, bukan angka)
lastMonth = (datetime.today().replace(day=1) - timedelta(days=1)).strftime("%B")

# Nama file dengan tanggal hari ini
today = datetime.today().strftime('%d-%m-%Y')
filename = f"Simpanan UKLN Bulanan {today}.xlsx"

# Daftar dataframe
dataframes = [D1, D2, D3, D4, D5, D6, D7, D8, D9, D10,
              D11, D12, D13, D14, D15, D16, D17, D18, D19, D20,
              D21, D22, D23, D24, D25, D26, D27, D28, D29, D30, D31]

# Gabungkan semua DataFrame, hilangkan baris kosong
df_combined = pd.concat([df for df in dataframes if not df.empty], ignore_index=True)

# Simpan ke Excel dengan auto-fit kolom
with pd.ExcelWriter(filename, engine="xlsxwriter") as writer:
    # Simpan masing-masing sheet
    for i, df in enumerate(dataframes, start=1):
        sheet_name = f"{i} {lastMonth}"
        df.to_excel(writer, sheet_name=sheet_name, index=False)

        # Auto-adjust column width
        worksheet = writer.sheets[sheet_name]
        for j, column in enumerate(df.columns):
            column_width = max(df[column].astype(str).map(len).max(), len(column))
            worksheet.set_column(j, j, column_width + 2)

    # Simpan sheet ringkasan
    df_combined.to_excel(writer, sheet_name="Summary", index=False)

    # Auto-fit kolom untuk sheet Summary
    worksheet = writer.sheets["Summary"]
    for j, column in enumerate(df_combined.columns):
        column_width = max(df_combined[column].astype(str).map(len).max(), len(column))
        worksheet.set_column(j, j, column_width + 2)
```
