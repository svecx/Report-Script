```python
import pandas as pd
from datetime import datetime, timedelta
from openpyxl import load_workbook
!pip install XlsxWriter
import calendar
from dateutil.relativedelta import relativedelta
```

```python
# def load_data():
#   try:

#   except Exception as e:
#     print(f"dae")

today = datetime.today()
prev_month = today - relativedelta(months=1)
base_date = datetime.strptime(prev_month.strftime("%d-%b-%y"), "%d-%b-%y")
#

# Format dasar tanggal
# base_date = datetime.strptime("01-Aug-25", "%d-%b-%y")

# Baca file dan filter data secara otomatis
FDataframes = []
for i in range(1, 32):
    try:
        D = pd.read_excel(f"/content/{i}.XLSX")
        FD = D[~D['Comp. Code'].isna()]  # hapus row kosong
        tanggal = (base_date.replace(day=i)).strftime("%d-%b-%y")
        FD["Tanggal"] = tanggal
        FDataframes.append(FD)
    except FileNotFoundError:
        print(f"File {i}.XLSX tidak ditemukan, dilewati.")

# Gabungkan semua DataFrame
df_combined = pd.concat(FDataframes, ignore_index=True)

# Hitung bulan lalu (nama bulan, bukan angka)
lastMonth = (datetime.today().replace(day=1) - timedelta(days=1)).strftime("%B")

# Nama file dengan tanggal hari ini
today = datetime.today().strftime('%d-%m-%Y')
filename = f"Simpanan UKLN Bulanan {today}.xlsx"

# Simpan ke Excel dengan auto-fit kolom
with pd.ExcelWriter(filename, engine="xlsxwriter") as writer:
    # Simpan sheet ringkasan
    df_combined.to_excel(writer, sheet_name="Summary", index=False)

    # Simpan masing-masing sheet
    for i, df in enumerate(FDataframes, start=1):
        sheet_name = f"{i} {lastMonth}"
        df.to_excel(writer, sheet_name=sheet_name, index=False)

        # Auto-adjust column width untuk tiap sheet
        worksheet = writer.sheets[sheet_name]
        for j, column in enumerate(df.columns):
            column_width = max(df[column].astype(str).map(len).max(), len(column))
            worksheet.set_column(j, j, column_width + 2)

    # Auto-fit kolom untuk sheet Summary
    worksheet = writer.sheets["Summary"]
    for j, column in enumerate(df_combined.columns):
        column_width = max(df_combined[column].astype(str).map(len).max(), len(column))
        worksheet.set_column(j, j, column_width + 2)
```
